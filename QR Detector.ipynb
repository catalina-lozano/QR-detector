```python
# Constants
WINDOW_NAME = 'QR Code Detector'
FONT = cv2.FONT_HERSHEY_SIMPLEX
FONT_SCALE = 1
FONT_COLOR = (255, 255, 255)
RECTANGLE_COLOR = (0, 255, 0)
RECTANGLE_THICKNESS = 2
TEXT_POSITION = (30, 100)
ESC_KEY = 27
```

```python
def draw_rectangle_with_text(img, points, text):
    # Print points for debugging
    print("Points structure:", points)

    x1, y1 = points[0]
    x2, y2 = points[2]
    
    # Draw rectangle
    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), RECTANGLE_COLOR, RECTANGLE_THICKNESS)
    
    # Draw text
    cv2.putText(img, text, TEXT_POSITION, FONT, FONT_SCALE, FONT_COLOR, 2)
```

```python
def main():
    
    # Set webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    # QR Code detector
    detector = cv2.QRCodeDetector()

    while True:
        success, img = cap.read()
        if not success:
            print("Error: Could not read frame.")
            break

        value, points, _ = detector.detectAndDecode(img)

        if points is not None:
            points = points[0]  # Extract the first (and only) set of points
            draw_rectangle_with_text(img, points, value)
            print(f"QR Code detected: {value}")
            print(f"Points: {points}")

        cv2.imshow(WINDOW_NAME, img)
        
        # Check if the window was closed
        if cv2.getWindowProperty(WINDOW_NAME, cv2.WND_PROP_VISIBLE) < 1:
            break

        if cv2.waitKey(1) & 0xFF == ESC_KEY:
            break

    cap.release()
    cv2.destroyAllWindows()  # Close all windows when the program exits

if __name__ == "__main__":
    main()
```
